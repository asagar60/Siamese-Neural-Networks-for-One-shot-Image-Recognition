## Plot Training Accuracy and Training Loss

```python
def plot_metric(loss, acc):

    x = np.array(range(1,1290))
    xnew = np.linspace(1,1289,100)

    spl = make_interp_spline(x, acc, k=3) #BSpline object
    ynew = spl(xnew)
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    plt.plot(acc,color = 'b', alpha=0.4, label = 'Avg Batch Accuracy')
    plt.plot(xnew, ynew,color = 'b', alpha=1, label = 'Smooth Average Accuracy')
    plt.xlabel('Steps (for every 500)')
    plt.ylabel('Accuracy')
    plt.legend(loc = "upper left")
    plt.title('Accuracy')

    spl = make_interp_spline(x, loss, k=3) #BSpline object
    ynew = spl(xnew)
    plt.subplot(1,2,2)
    plt.plot(loss,color = 'b', label = 'Avg Batch Loss')
    #plt.plot((xnew, ynew),color = 'b', alpha=1)
    plt.xlabel('Steps (for every 500)')
    plt.ylabel('Loss')
    plt.legend(loc = "upper left")
    plt.title('Loss')
```

## Testing N- way one shot 

### Nearest Neighbour model as base model

```python
def nearest_neighbour_correct(pairs,targets):
    """returns 1 if nearest neighbour gets the correct answer for a one-shot task
        given by (pairs, targets)"""
    X_left, X_right = pairs
    L2_distances = np.zeros_like(targets)
    for i in range(len(targets)):
        L2_distances[i] = np.sqrt(abs(np.sum(X_left[i]**2 - X_right[i]**2)))
    if np.argmin(L2_distances) == np.argmax(targets):
        return 1
    return 0
```

### Defining Function for Within Alphabet and Unstructured Alphabet Testing

```python
def within_alphabet_test_pairs(n_way = 20):

    correct_pred = 0
    nn_correct =  0
    alphabets = [item[0] for item in alphabet.items()]
    img_paths= []
    for a in alphabets:
        img_paths.append([items[0] for items in image_data.items() if items[1][0] == a])
    
    total_run = 0
    for a in alphabets:
        set_1, set_2 = [], []
        
        sampled_characters = random.sample(list(alphabet[a]),n_way)
        for character in sampled_characters:
            char_paths = [items[0] for items in image_data.items() if items[1][1] == character and items[1][0]==a]
            choosen_imgs = random.sample(char_paths,2)
            set_1.append(choosen_imgs[0])
            set_2.append(choosen_imgs[1])
        
        #20 - examples
        
        for i, img0_path in enumerate(set_1):
            X_left, X_right,y = [],[],[]
            for j,img1_path in enumerate(set_2):
                #print(img0)
                img0 = mpimg.imread(img0_path)
                img0 = np.expand_dims(img0, axis = 2)
                img1 = mpimg.imread(img1_path)
                img1 = np.expand_dims(img1, axis = 2)
                X_left.append(img0)
                X_right.append(img1) 
                y.append(1 if i==j else 0)
        
            X_left, X_right = np.array(X_left), np.array(X_right)
            correct_pred += test_one_shot(X_left, X_right, y)
            nn_correct += nearest_neighbour_correct((X_left, X_right), y)
            total_run +=1
        #20 - examples
        
        for i, img0_path in enumerate(set_2):
            X_left, X_right,y = [],[],[]
            for j,img1_path in enumerate(set_1):
                img0 = mpimg.imread(img0_path)
                img0 = np.expand_dims(img0, axis = 2)
                img1 = mpimg.imread(img1_path)
                img1 = np.expand_dims(img1, axis = 2)
                X_left.append(img0)
                X_right.append(img1) 
                y.append(1 if i==j else 0)
        
            X_left, X_right = np.array(X_left), np.array(X_right)
            correct_pred += test_one_shot(X_left, X_right, y)
            nn_correct += nearest_neighbour_correct((X_left, X_right), y)
            total_run +=1 
    #total_run = len(alphabets * n_way * 2)
    return (correct_pred*100/total_run, nn_correct*100/total_run)
```

```python
def unstructured_test_pairs(n_way = 20):
    
    correct_pred = 0
    nn_correct = 0
    alphabets = [item[0] for item in alphabet.items()]
    img_paths= []
    for a in alphabets:
        img_paths.append([items[0] for items in image_data.items() if items[1][0] == a])
    
    unique_label_list = []
    for items in alphabet.items():
        for item in items[1]:
            unique_label_list.append([items[0], item])
        
    img_paths = []
    for label in unique_label_list:
        img_paths.append([items[0] for items in image_data.items() if items[1] == label])
        
    total_img_pairs = 0
    
    for i in tqdm(range(len(unique_label_list))):

        X_left, X_right,y = [],[],[]        
        im_paths = img_paths[i]
        img0_1 = random.sample(im_paths,2)

        img0 = mpimg.imread(img0_1[0])
        img0 = np.expand_dims(img0, axis = 2)
        img1 = mpimg.imread(img0_1[1])
        img1 = np.expand_dims(img1, axis = 2)
        X_left.append(img0)
        X_right.append(img1)
        y.append(1)

        for k in range(n_way-1):
            random_pick = random.sample(image_data.items(),1)
            while (random_pick[0][1] == unique_label_list[i]):
                random_pick = random.sample(image_data.items(),1)
            img2_path = random_pick[0][0]
            img2 = mpimg.imread(img2_path)
            img2 = np.expand_dims(img2, axis = 2)
            X_left.append(img0)
            X_right.append(img2)
            y.append(0)

        X_left, X_right = np.array(X_left), np.array(X_right)
        correct_pred += test_one_shot(X_left, X_right, y)
        nn_correct += nearest_neighbour_correct((X_left, X_right), y)
        
    
    total_run = len(unique_label_list)
    return (correct_pred*100/total_run, nn_correct*100/total_run)
```

```python
def test_validation_acc():
    wA_acc = within_alphabet_test_pairs()
    u_Acc = unstructured_test_pairs()
    return [wA_acc, u_Acc]
```

```python
def test_one_shot(X_left,X_right, y):
    model = siamese_net
    prob = model.predict([X_left,X_right])
    if np.argmax(prob) == np.argmax(y):
        return 1
    else:
        return 0
```

## Evaluation on 10 Alphabets ( Evaluation Set)

```python
file_name = 'eval_im_paths_10.pkl'

with open(file_name, 'rb') as f:
    image_data, alphabet = pkl.load(f)
```

```python
def one_shot_accuracy():
    within_accuracies = []
    unstructred_accuracies = []
    #[2, 5, 6, 10, 15, 16, 20]
    for i in range(2, 21):
        within_accuracies.append(within_alphabet_test_pairs(n_way = i))
        unstructred_accuracies.append(unstructured_test_pairs(n_way = i)) 
    return within_accuracies ,unstructred_accuracies
```

```python
within_accuracies ,unstructred_accuracies = one_shot_accuracy()
```

```python
for i, acc in enumerate(within_accuracies):
    print('For {}-way accuracy - Oneshot: {:.2f} , Nearest Neighbor: {:.2f}'.format(i+2, acc[0], acc[1]))
```

```python
for i, acc in enumerate(unstructred_accuracies):
    print('For {}-way accuracy - Oneshot: {:.2f} , Nearest Neighbor: {:.2f}'.format(i+2, acc[0], acc[1]))
```

```python
ways = np.arange(2, 21, 1)

plt.plot(ways,np.asarray(within_accuracies)[:,0], color = 'b',  label = 'wA 20-way one shot accuracy')
plt.plot(ways,np.asarray(unstructred_accuracies)[:,0], color = 'r', label = 'uA 20-way one shot accuracy')
plt.plot(ways,np.asarray(within_accuracies)[:,1], color = 'g',  label = 'wA 20-way nearest neighbour accuracy')
plt.plot(ways,np.asarray(unstructred_accuracies)[:,1], color = 'm', label = 'uA 20-way nearest neighbour accuracy')
plt.xlabel('Steps (for every 500)')
plt.xticks(np.arange(2, 22, step=2))
plt.ylabel('N-way Accuracy')
plt.legend(loc = "bottom left")
plt.title('N-way Accuracies')
```
